## 测试cuda是否可用

```python
import torch
print(torch.cuda.is_available())  # Should be True
print(torch.cuda.device_count())  # Should be > 0
```

```python
import torch
print(torch.__version__)
print(torch.cuda.is_available())
import d2l
print(d2l.__version__)
```

```python
import torch
print(torch.__version__)  # 应该类似 2.6.0.dev20251017
print(torch.cuda.is_available())  # 应该为 True
print(torch.cuda.get_device_name(0))  # 应该为 RTX 5060 Ti
```

```python
import torch, sys, os
print("PyTorch version :", torch.__version__)          # 必须含 **dev** 或 **nightly**
print("CUDA runtime    :", torch.version.cuda)         # 应该 ≥ 12.6
print("cuDNN build     :", torch.backends.cudnn.version())
print("Python exe      :", sys.executable)
print("TORCH_CUDA_ARCH_LIST:", os.environ.get("TORCH_CUDA_ARCH_LIST", "未设置"))
```
